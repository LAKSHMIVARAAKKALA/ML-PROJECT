```python
#Storing the first three columns in x  and the last column in y
x=set.iloc[:,:-1]# all columns except the last row
y=set.iloc[:,-1]

#Displaying the results
print("Features (x):")
print(x)
print("\n target (y):")
print(y)
```

```python
object_columns = x.select_dtypes(include=['object']).columns
print(object_columns)
```

```python
# We should handil the missing values but there are no missing values in the set(dataset)
from sklearn.preprocessing import LabelEncoder


#Create a LabelEncoder instance
label_encoder=LabelEncoder()

#List of catagorical columns to encode
categorical_columns=object_columns

#loop through each categorical column and apply labelencoding
for column in categorical_columns:
  x[column]=label_encoder.fit_transform(x[column])

# Display the updated DataFrame
print("\nUpdated Features (x) after Label Encoding:")
print(x)
```

```python
y=label_encoder.fit_transform(y)
print("\nUpdated Features (y) after Label Encoding:")
print(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
x_test_scaled
```

```python
x_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
LogisticRegression_model=LogisticRegression(random_state=0)
LogisticRegression_model.fit(x_train_scaled,y_train)
```

```python
st_prediction=LogisticRegression_model.predict(x_test_scaled)
print("LogisticRegresstion:")
print("Accuracy:",accuracy_score(y_test,st_prediction))
```

```python
print(classification_report(y_test,st_prediction))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
student=DecisionTreeClassifier(random_state=0)
student.fit(x_train_scaled,y_train)
```

```python
std_prediction=student.predict(x_test_scaled)
print("DecisionTreeClassifier:")
print("Accuracy:",accuracy_score(y_test,std_prediction))
```

```python
print(classification_report(y_test,std_prediction))
```

```python
stu=RandomForestClassifier(random_state=0)
stu.fit(x_train_scaled,y_train)
st_prediction=stu.predict(x_test_scaled)
print("RandomForestClassifier:")
print("Accuracy:",accuracy_score(y_test,st_prediction))
```

```python
print(classification_report(y_test,st_prediction))
```

```python
# Predict on the test data
y_pred = stu.predict(x_test_scaled)
print("RandomForestClassifier report:")
print(classification_report(y_test, y_pred))
```
